<h1>Helper Functions </h1> 

```python
'''
The switch method takes in a midi note as an argument,
queries the switcher array and returns the image associated with the corresponding note.

If the midi_note is not assigned, it will return the default image at [126]

'''
def switch(midi_note):
     
    if midi_note == '':
        return (switcher[126])
    
    else:
        y = int(midi_note)
        if switcher [y] == '':
            return switcher[126]
        else:
            return (switcher[y])
    

'''
The brightness effect boosts the luminance of the pixel by the value brightness_val.
There is a limit of 255 - if the bightness value + the current color value > 255: set
value to 255.
'''
    
def fx_bright(color_val, bright_val):
    
    clip = 255
    z = color_val + bright_val
    
    if z > 255: 
        return clip
   
    else: 
        return z

    
'''
The Contrast function boosts the color by the contrast_val amount if it is above the pivot value (127)    
and cuts the value if below the pivot value. 
This function increases the difference between light and dark colors.
'''
def fx_contrast(color_val, contrast_val):
   
    out = 0
    
    if color_val > 127:
        
        out = color_val + contrast_val
        
        if out > 254:
            out = 254
        
        else:
            out = out
        
    elif color_val <= 127:
        
        out = color_val - contrast_val
        
        if out < 0:
            out = 0
        
        else:
            out = out
        
    return out

'''
The Noir function takes in all three rgb values and returns brightness value by averaging the three color channels. 
This has the effecet of converting the image to black and white .

We can also use the Luminance formula to return a more defined black and white image. To select the luminance method,
pass a True as the fourth variable
'''

def fx_noir(r, g, b):
    avg = (r + b + g) / 3
    return avg

def fx_noir(r, g, b, lum_true):
    if lum_true: 
        lum = (.21 * r) + (.72 * b) + (g * .07)     
        return lum
    else:
        return fx_noir(r, g, b)

def fx_sat(r, g, b):
    pass
    
```

<h1>Main PyGame Instance</h1>

```python
import numpy as np
from PIL import Image
import pygame
import pygame.midi
import time
import random 
from math import *

#midi channel which pygame will listen for inputs 
midi_channel = 1

#Size of the board
width = 1000
height = 1000

#Size of the pixel
w = 3
h = 3
edge = 1

#Image effects -- 1 == on , 0 == off
bright = 0
contrast = 0
noir = 0
sat = 0

img_width = width
img_height = height


# initialize pygame window
pygame.init()
pygame.midi.init()
midi_in = pygame.midi.Input(midi_channel) # --!important! -- this
screen = pygame.display.set_mode([img_width,img_height])

midi_note = ''


running = True
while running:
    
    #Read in 60 samples from the midi buffer
    stack = midi_in.read(60)
    
    
    #Seperate valid midi notes from invalid notes in the buffer --- 144 = note start --- 126 = note finish
    if len(stack) > 0:
        
        
        io = 0
        while io < len(stack): 
            if stack[io][0][0] == 144:
                midi_note = stack[0][0][1]
                #print(midi_note)
            
            elif stack[io][0][0] == 128:
                #print('end')
                midi_note = '126'
            
            else:
                midi_note = midi_note
            
            io = io + 1
           
    midi_note = midi_note
            
        
    #resize image to fit board
    pre_size = Image.open(switch(midi_note))
    resize = pre_size.resize((img_width, img_height))
    

    #load pixel data to array (r,g,b)
    pix_val = resize.load()
       
    #FIX ME link this sleep to the bpm of the midi track?
    time.sleep(.0001)
     
    r = 0
    g = 0
    b = 0
        
    # Handle exit of window    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    
    #Background color
    screen.fill((0,0,0))
    ypos = 0
    
    #Traverse pix_val array to retrieve pixel data for each node 
    while ypos < img_height:
        xpos = 0
       
            
        while xpos < img_width:
            
            #Base rgb data per pixel
            r = pix_val[xpos, ypos][0]
            g = pix_val[xpos, ypos][1]
            b = pix_val[xpos, ypos][2]
            
            #Image FX function calls
            if bright > 0:
               
                r = fx_bright(r, bright)
                g = fx_bright(g, bright)
                b = fx_bright(b, bright)
            
            if noir == 1:
                
                r = fx_noir(r, g, b)
                g = r
                b = r
                
            if contrast > 0: 
               
                r = fx_contrast(r, contrast)
                g = fx_contrast(g, contrast)
                b = fx_contrast(b, contrast)
            
            if sat > 0:
                r = sat(r, g, b, sat)
                g = sat(g, b, r, sat)
                b = sat(b, r, g, sat)
            
            #draw Pixels to screen
            pygame.draw.rect(screen,(r, g, b), (xpos , ypos ,w - edge , h - edge), 0)
            
            xpos = xpos + w
        
        ypos = ypos + h
    
    pygame.display.flip()

midi_in.close()
pygame.midi.quit()    
pygame.quit()
```
